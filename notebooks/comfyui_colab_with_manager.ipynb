```python
import json
import requests
import time
from openai import OpenAI # 用于 DeepSeek API

# --- 配置区 ---
DEEPSEEK_API_KEY = "sk-b5a62da200b34d30b0a9c24a9f0bc378"
COMFY_URL = "http://127.0.0.1:8188"
PROMPT_NODE_ID = "6"  # 你的 CLIPTextEncode 节点 ID

# 初始化 DeepSeek 客户端
deepseek_client = OpenAI(api_key=DEEPSEEK_API_KEY, base_url="https://api.deepseek.com/v1") # 请确认 DeepSeek 的 base_url

def get_detailed_scripts_from_deepseek(story_theme):
    """
    调用 DeepSeek 生成包含分镜原文和绘画 Prompt 的剧本。
    返回格式：[{"text": "中文台词", "prompt": "英文绘画Prompt"}, ...]
    """
    prompt = f"""
    请根据主题“{story_theme}”写一段短篇故事，并将其拆解为3个视觉分镜。
    每个分镜需要包含：
    1. 【text】: 对应的中文台词（最多30字）。
    2. 【prompt】: 详细的英文绘画提示词(Prompt)，要求包含画质词、风格、光影和人物神态描述，并且人物角色形象保持一致。

    输出格式要求为纯 JSON 数组，例如：
    [
      {{"text": "程序员李明在霓虹闪烁的未来城市中，眼神疲惫却坚定。", "prompt": "A young Asian male programmer, named Li Ming, with tired but determined eyes, standing in a neon-lit cyberpunk city, detailed, cinematic, high contrast, rainy street, blue and purple lighting, futuristic outfit, sharp focus, full shot"}},
      {{"text": "他成功写完AI助手代码，屏幕上弹出“财富自由”的字样。", "prompt": "Li Ming, the programmer, sitting in front of a holographic screen displaying 'Financial Freedom' in bright green text, his face showing a mix of exhaustion and triumph, warm glow from screen, highly detailed, indoor scene, soft light"}},
      {{"text": "李明在海边别墅，享受着自由与宁静，夕阳洒在他的脸上。", "prompt": "Li Ming, the programmer, now dressed in elegant casual wear, relaxing on a balcony of a luxurious futuristic villa overlooking a calm ocean at sunset, peaceful expression, golden hour light, highly detailed, serene atmosphere"}}
    ]
    """

    try:
        response = deepseek_client.chat.completions.create(
            model="deepseek-chat", # 或 deepseek-coder
            messages=[{"role": "user", "content": prompt}],
            response_format={"type": "json_object"}
        )
        content_str = response.choices[0].message.content
        scripts = json.loads(content_str)
        return scripts if isinstance(scripts, list) else []
    except Exception as e:
        print(f"❌ DeepSeek API 调用或解析错误: {e}")
        return []

# --- 辅助函数：将 Prompt 注入 JSON 并发送给 ComfyUI (无需修改) ---
def run_comfy_task(english_prompt):
    with open("workflow_api.json", "r") as f:
        workflow = json.load(f)
    workflow[PROMPT_NODE_ID]["inputs"]["text"] = english_prompt
    requests.post(f"{COMFY_URL}/prompt", json={"prompt": workflow})

# --- （可选）如果你想在这里等待图片生成完成再继续，需要更复杂的逻辑，目前我们先让它异步 ---
# ComfyUI 的 API 默认是异步加入队列，图片生成需要时间。
# 如果需要同步等待，需要监听 ComfyUI 的 websocket 接口，暂时跳过。

# --- 执行主流程 (保留 ComfyUI 启动后在单独 Cell 运行的部分) ---
# 注意：你需要确保你的 ComfyUI 已经在另一个 Cell 启动了。
# 这个 Cell 专注于 DeepSeek 生成剧本和 ComfyUI 任务发送。

# if __name__ == "__main__": # 如果你将这段代码放在独立脚本中运行
theme = "一名程序员在 2026 年通过 AI 副业获得财富自由的赛博朋克故事"
print(f"🎬 正在让 DeepSeek 构思剧本: {theme}...")

try:
    detailed_scripts = get_detailed_scripts_from_deepseek(theme)
    print(f"✅ 剧本生成成功，准备绘制 {len(detailed_scripts)} 张分镜图。")

    # 存储生成的图片文件名，以便后续合成视频
    generated_image_paths = []

    # 假设图片会以某种方式命名，例如 `image_00001_.png`。
    # 实际 ComfyUI 的文件名生成逻辑比较复杂，我们先假设它按顺序生成到 output 目录。
    # 为了简化，我们假设ComfyUI每次生成图片都叫 "ComfyUI_00001_.png" 然后递增
    # 你可能需要手动修改ComfyUI的节点，让它以 `index_yourprompt.png` 格式输出

    for i, script_data in enumerate(detailed_scripts):
        print(f"🎨 正在绘制第 {i+1} 张: {script_data['prompt'][:50]}...")
        run_comfy_task(script_data['prompt'])

        # 暂时等待几秒，让 ComfyUI 有时间生成图片
        # 实际项目中，你需要监听 ComfyUI 的 websocket 来确认图片生成完成
        time.sleep(10) # 增加等待时间，确保图片有足够时间生成

        # 假设 ComfyUI 每次生成都会保存到 /content/drive/MyDrive/ComfyUI/output
        # 并且文件名是 ComfyUI_0000X_.png 这样的格式，我们需要最晚生成的那张
        # 这是一个简化的处理，实际可能需要更严谨的文件监控
        output_dir = "/content/drive/MyDrive/ComfyUI/output"
        # 找出最新生成的文件
        latest_file = max([f for f in os.listdir(output_dir) if f.endswith('.png')],
                           key=lambda x: os.path.getmtime(os.path.join(output_dir, x)))
        latest_path = os.path.join(output_dir, latest_file)
        generated_image_paths.append(latest_path)
        print(f"📸 已获取图片路径: {latest_path}")

    print("\n✨ 所有绘图任务已加入队列！请刷新 ComfyUI 网页查看进度，或等待图片在 Colab 文件夹中出现。")
    print(f"所有生成的图片路径: {generated_image_paths}")

except Exception as e:
    print(f"❌ 运行失败: {e}")
```

```python
import asyncio
import edge_tts
import os

# 定义语音输出目录
AUDIO_OUTPUT_DIR = "audio_clips"
os.makedirs(AUDIO_OUTPUT_DIR, exist_ok=True)

async def generate_audio_for_scripts(scripts_data):
    """为每个剧本分镜生成音频文件"""
    audio_paths = []
    for i, script_data in enumerate(scripts_data):
        text_to_speak = script_data['text']
        output_file = os.path.join(AUDIO_OUTPUT_DIR, f"audio_{i+1}.mp3")

        # 选择一个中文语音（例如：zh-CN-XiaoxiaoNeural 是女声）
        # 更多声音请参考：https://github.com/rany2/edge-tts#voices
        voice = "zh-CN-XiaoxiaoNeural" # 或者 zh-CN-YunyangNeural (男声)

        print(f"🎤 正在为分镜 {i+1} 生成音频: '{text_to_speak}'...")
        communicate = edge_tts.Communicate(text_to_speak, voice)
        await communicate.save(output_file)
        audio_paths.append(output_file)
        print(f"✅ 音频 '{output_file}' 生成完成。")
    return audio_paths

# --- 在主流程的 ComfyUI 任务完成后调用 ---
if 'detailed_scripts' in locals() and detailed_scripts: # 确保剧本已生成
    print("\n🎧 正在生成音频...")
    # asyncio.run() 只能在最顶层调用一次
    # 如果你已经在 Colab 运行，需要把这个包装一下
    audio_file_paths = []

    # Colab 环境下，直接用 run() 会报错，需要用 nest_asyncio
    import nest_asyncio
    nest_asyncio.apply()

    audio_file_paths = asyncio.run(generate_audio_for_scripts(detailed_scripts))
    print(f"所有生成的音频路径: {audio_file_paths}")
```

```python
from moviepy.editor import ImageClip, AudioFileClip, concatenate_videoclips

VIDEO_OUTPUT_DIR = "final_videos"
os.makedirs(VIDEO_OUTPUT_DIR, exist_ok=True)

def create_video_from_clips(image_paths, audio_paths, output_filename="final_manga_video.mp4"):
    """
    将图片和对应的音频合成视频。
    每张图片的持续时间由其对应的音频长度决定。
    """
    if len(image_paths) != len(audio_paths) or not image_paths:
        print("❌ 错误：图片和音频数量不匹配，或没有可用的素材。")
        return None

    clips = []
    for i in range(len(image_paths)):
        img_path = image_paths[i]
        aud_path = audio_paths[i]

        # 创建图片剪辑
        img_clip = ImageClip(img_path)

        # 加载音频剪辑，并获取其时长
        audio_clip = AudioFileClip(aud_path)

        # 设置图片剪辑的时长与音频时长一致
        img_clip = img_clip.set_duration(audio_clip.duration)

        # 将音频添加到图片剪辑中
        video_clip = img_clip.set_audio(audio_clip)

        clips.append(video_clip)

    print(f"\n🎥 正在合并 {len(clips)} 个视频片段...")
    final_clip = concatenate_videoclips(clips, method="compose")

    output_path = os.path.join(VIDEO_OUTPUT_DIR, output_filename)
    final_clip.write_videofile(output_path, fps=24, codec="libx264") # 可以调整 fps
    print(f"✅ 视频生成成功，保存至: {output_path}")
    return output_path

# --- 在主流程的音频生成完成后调用 ---
if 'generated_image_paths' in locals() and generated_image_paths and \
   'audio_file_paths' in locals() and audio_file_paths:
    final_video_path = create_video_from_clips(generated_image_paths, audio_file_paths)
    if final_video_path:
        print(f"\n🎉 恭喜！你的 AI 漫剧视频已完成：{final_video_path}")
```
